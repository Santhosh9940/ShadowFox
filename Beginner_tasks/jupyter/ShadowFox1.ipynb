```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [100, 120, 115, 150, 170]

plt.plot(x, y, color='blue')
plt.title("Monthly Sales Over Time")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid()
plt.show()
```

```python
#Bar Plot
import matplotlib.pyplot as plt

departments = ['HR', 'Sales', 'IT', 'Finance']
employees = [10, 25, 15, 20]

plt.bar(departments, employees, color='green')
plt.title("Employees per Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

```python
#Horizontal bar plot
import matplotlib.pyplot as plt

languages = ['Python', 'Java', 'C++', 'Ruby']
popularity = [80, 70, 65, 30]

plt.barh(languages, popularity, color='purple')
plt.title("Programming Language Popularity")
plt.xlabel("Popularity")
plt.ylabel("Language")
plt.show()
```

```python
#Histogram
import matplotlib.pyplot as plt
import numpy as np

scores = np.random.normal(70, 10, 200)

plt.hist(scores, bins=10, color='orange', edgecolor='black')
plt.title("Score Distribution")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```

```python
#scatterplot
import matplotlib.pyplot as plt

age = [22, 25, 47, 52, 46, 56, 48, 35]
income = [25000, 27000, 45000, 50000, 48000, 52000, 47000, 40000]

plt.scatter(age, income, color='red')
plt.title("Age vs Income")
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()
```

```python
#piechart
import matplotlib.pyplot as plt

labels = ['Rent', 'Food', 'Transport', 'Others']
sizes = [35, 30, 20, 15]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Monthly Budget")
plt.axis('equal')
plt.show()
```

```python
#boxplot
import matplotlib.pyplot as plt
import numpy as np

data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data, patch_artist=True)
plt.title("Distributions")
plt.xlabel("Sample Group")
plt.ylabel("Values")
plt.show()
```

```python
#pandas
import pandas as pd

marks = pd.Series([85, 90, 78, 92], index=['Alice', 'Bob', 'Charlie', 'David'])

print(marks)
```

```python
#head and tail
s = pd.Series([10, 20, 30, 40, 50, 60])
print(s.head(3))
print(s.tail(2))
```

```python
#sum,mean,median,std
s = pd.Series([100, 200, 300, 400])
print("Sum:", s.sum())
print("Mean:", s.mean())
print("Median:", s.median())
print("Standard Deviation:", s.std())
```

```python
#value_counts
s = pd.Series(['apple', 'banana', 'apple', 'orange', 'banana', 'banana'])
print(s.value_counts())
```

```python
#mapping
s = pd.Series(['cat', 'dog', 'bird'])
mapping = {'cat': 1, 'dog': 2, 'bird': 3}
print(s.map(mapping))
```

```python
#DataFrame example
import pandas as pd

data = {
    'Department': ['HR', 'HR', 'IT', 'IT', 'Finance', 'Finance'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'Salary': [3000, 3500, 4000, 4200, 5000, 5200],
    'Experience': [2, 3, 5, 4, 6, 7]
}

df = pd.DataFrame(data)

agg_result = df.groupby('Department').agg({
    'Salary': ['sum', 'mean', 'max', 'min', 'std'],
    'Experience': ['mean', 'count', 'max']
})

print(agg_result)
```

```python
#isnull
df = pd.DataFrame({
    'Name': ['Alice', None, 'Charlie'],
    'Age': [25, 30, None]
})

print(df.isnull())
```

```python
#fillna()
df_filled = df.fillna({'Nmae': 'Unknown', 'Age': df['Age'].mean()})
print(df_filled)
```

```python
#dropna()
print(df.dropna())
print(df.dropna(axis=1))
```
